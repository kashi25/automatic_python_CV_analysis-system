```python
import os
import csv
import pdfplumber
import re

# Define the folder path
folder_path = 'E:/csv'

# Define the target words and sentences you want to search for
target_words = ["RESTAPI", "python", "data", "git", "html", "css", "mysql", "CSS", "java", "django"]
target_sentences = ["experience", "junior", "mid","Internships","entrylevel"]

# Define the regular expression patterns for email addresses and phone numbers
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
phone_pattern = r'\b(\d{3})[-.]?(\d{3})[-.]?(\d{4})\b'

# Create a list to store the extracted information
results = []

# Iterate over the PDF files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        # Open the PDF file
        pdf_path = os.path.join(folder_path, filename)
        pdf = pdfplumber.open(pdf_path)
        
        # Extract the PDF name
        pdf_name = os.path.basename(pdf_path)
        
        # Initialize variables for storing matching words and sentences
        matching_words = []
        matching_sentences = []
        email_addresses = []
        contact_numbers = []

        # Iterate through each page of the PDF
        for page in pdf.pages:
            # Extract text from the page and convert to lowercase
            text = page.extract_text().lower()

            # Search for the target words in the text (case-insensitive)
            for word in target_words:
                if word.lower() in text:
                    matching_words.append(word)

            # Search for the target sentences in the text (case-insensitive)
            for sentence in target_sentences:
                if sentence.lower() in text:
                    matching_sentences.append(sentence)

            # Find email addresses in the text using regular expressions
            emails = re.findall(email_pattern, text)
            email_addresses.extend(emails)

            # Find phone numbers in the text using regular expressions
            phones = re.findall(phone_pattern, text)
            contact_numbers.extend(phones)

        # Close the PDF file
        pdf.close()

        # Calculate the total number of matching words
        total_matching_words = len(matching_words)

        # Store the extracted information in a dictionary
        result = {
            'PDF Name': pdf_name,
            'Matching Words': matching_words,
            'Matching_sentences': matching_sentences,
            'Total Matching Words': total_matching_words,
            'Email Addresses': email_addresses,
            'Contact Numbers': contact_numbers
        }

        # Append the result to the list
        results.append(result)

# Sort the results based on the total matching words in descending order
results.sort(key=lambda x: x['Total Matching Words'], reverse=True)

# Define the output CSV file path
output_file = os.path.join(folder_path, 'kashi_output.csv')

# Write the results to the CSV file
with open(output_file, 'w', newline='') as file:
    fieldnames = ['PDF Name', 'Matching Words','Matching_sentences' ,'Total Matching Words', 'Email Addresses', 'Contact Numbers']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(results)

print("Extraction completed. Results saved in:", output_file)
```
